```python
from skimage.exposure import adjust_sigmoid

def preprocess(img):
    gray =cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    new=adjust_sigmoid(gray, cutoff=0.6, gain=5, inv=False)
    kernel = np.ones((5,5), np.uint8)
#     new = cv2.dilate(new, kernel, iterations=1)
    stacked_img = np.stack((new,)*3, axis=-1)
    return stacked_img
```

```python
img=io.imread('test/examples/test2.jpg')
preprocessed=preprocess(img)
plt.imshow(preprocessed)
```

```python
cv2.imwrite('test/preprocessed/test2.jpg',preprocessed)
```
